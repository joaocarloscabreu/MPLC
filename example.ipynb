```python
from MPLC.system import System
import matplotlib.pyplot as plt
import numpy as np

s = System(iterations=200)
```

```python
Gaussian = s.create_Gaussians(num_modes=5, positions="triangle",factor_mult=0.8)
```

```python
fig,ax = Gaussian.plot()
```

```python

Hermite_plot =s.create_HermiteGaussian(w1 = Gaussian.w*4,  num_modes=5,plot=True, positions="triangle", rotation= 0, factor=1)
```

```python
fig,ax = Hermite_plot.plot(X=s.X, Y=s.Y)
```

```python
Hermite = s.create_HermiteGaussian()
```

```python
fig,ax = Hermite.plot(X=s.X, Y=s.Y)
```

```python
s.set_fields(Gaussian,Hermite)
```

```python
s.start()
```
